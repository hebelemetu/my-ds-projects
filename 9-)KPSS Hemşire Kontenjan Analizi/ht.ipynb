```python
yer = list(df["SEMT"])
anadoluda = list()
avrupada = list()
for x in yer:
    if x in anadolu_ilce:
        anadoluda.append("YES")
    else:
        anadoluda.append("NO")
for x in yer:
    if x in avrupa_ilce:
        avrupada.append("YES")
    else:
        avrupada.append("NO")
```

```python
df["ANADOLUDA"] = anadoluda
df["AVRUPADA"] = avrupada
df_new = df[df["ANADOLUDA"]=="YES"]
df_new = df_new.drop(["AVRUPADA","ANADOLUDA"],axis=1)
df_new.head()
```
