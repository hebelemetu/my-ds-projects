Bu veri setinde bağımlı değişken: memnuniyet_düzeyi seçilecektir. Diğer değişkenler bağımsız değişken olarak kullanılacak.

```python
#Kategorik olan özniteliklerin sayıya çevrilmesi
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
df["calistigi_bolum"]=labelencoder.fit_transform(df["calistigi_bolum"])
df["maas_derecesi"]=labelencoder.fit_transform(df["maas_derecesi"])
```

```python
df.head()
```

```python
#y bağımsız değişkeni X bağımlı değişkenleri ifade ediyor
X=df.drop("memnuniyet_duzeyi",axis=1).copy()
y = df["memnuniyet_duzeyi"].copy()
```

```python
#Öznitelik seçimi için en çok etkisi olan özniteliklerin kontrolü
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
from matplotlib import pyplot
def select_features(X_train, y_train, X_test):
    fs = SelectKBest(score_func=f_regression, k='all')
    fs.fit(X_train, y_train)
    X_train_fs = fs.transform(X_train)
    X_test_fs = fs.transform(X_test)
    return X_train_fs, X_test_fs, fs
```

```python
from sklearn.model_selection import train_test_split
# Train ve test dataset ayrımı %30 test % 70 train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
#Öznitelik seçimi
X_train_fs, X_test_fs, fs = select_features(X_train, y_train, X_test)
# Özniteliklerin değerlendirilmesi
for i in range(len(fs.scores_)):
    print('Feature %d: %f' % (i, fs.scores_[i]))
# Sonucun görselleştirilmesi
pyplot.bar([i for i in range(len(fs.scores_))], fs.scores_)
pyplot.show()
```

```python
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.svm import LinearSVR, SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
```

```python
#Regresyon modeli oluşturulması. Bunun için birçok regresyon modeli kullanılacak. İçlerinden en iyi model seçilecek.
models = {
    "                     Linear Regression": LinearRegression(),
    "                 L2 (Ridge) Regression": Ridge(),
    "Support Vector Machine (Linear Kernel)": LinearSVR(),
    "   Support Vector Mahcine (RBF Kernel)": SVR(),
    "                         Decision Tree": DecisionTreeRegressor(),
    "                        Neural Network": MLPRegressor(),
    "                         Random Forest": RandomForestRegressor(),
    "                     Gradient Boosting": GradientBoostingRegressor(),
    "                              AdaBoost": AdaBoostRegressor()
}

for name, model in models.items():
    model.fit(X_train, y_train)
    print(name + " trained.")
```

```python
#Her bir modelim tahmin skor değerlendirmesi
for name, model in models.items():
    print(name + " R^2: {:.5f}".format(model.score(X_test, y_test)))
```

```python
# Skoru en iyi modelin seçilmesi
best_model = GradientBoostingRegressor()
best_model.fit(X_train, y_train)

print("Model R^2 (Before Optimization): {:.5f}".format(best_model.score(X_test, y_test)))
```

```python
#Metriklerin optimize edilmesi
from sklearn.model_selection import GridSearchCV
params = {
    'learning_rate': [0.01, 0.1, 1.0],
    'n_estimators': [100, 150, 200],
    'max_depth': [3, 4, 5]
}

clf = GridSearchCV(best_model, params)
clf.fit(X_train, y_train)

clf.best_params_
```

```python
#Optimize edilmiş modelin tahmin skoru
print("Model R^2 (After Optimization): {:.5f}".format(clf.score(X_test, y_test)))
```

```python
#Farklı metrik ölçümü
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error
y_hat_test = clf.predict(X_test)
test_mse = mean_squared_error(y_test, y_hat_test)
test_rmse = np.sqrt(test_mse)
test_mae = mean_absolute_error(y_test, y_hat_test)
print(test_mse,test_rmse,test_mae)
```

```python
#Regresyon problemi için girdi tahmini
print(f"Regression tahmini  = {clf.predict(np.array([0.86,1,2]).reshape(1, 3))}")
```

```python
#Sınıflandırma problemi DecisisionTreeClassifier kullanılacak ve gridsearchcv kullanarak hyperparameter tuning yapılacak
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
X=df.drop("sirketten_ayrildi",axis=1).copy()
y = df["sirketten_ayrildi"].copy()
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
decisiontree = DecisionTreeClassifier(criterion='entropy')

params = {'max_depth': [4,5,6], 'min_samples_leaf': [1,2,3], 'min_samples_split': [2,3,4]}
grid_search = GridSearchCV(estimator=decisiontree, param_grid=params, cv=5)
grid_search.fit(X_train, y_train)
```

```python
#Parameterların sıralanmış hali
import pandas as pd
pd.DataFrame(grid_search.cv_results_).sort_values(by='rank_test_score')
```

```python
#Test datasına göre modelin metrik ölçümü
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
y_pred = grid_search.predict(X_test)
```

```python
acc = accuracy_score(y_test, y_pred)
pre = precision_score(y_test, y_pred, average='micro')
rec = recall_score(y_test, y_pred, average='micro')
f1 = f1_score(y_test, y_pred, average='micro')
acc, pre, rec, f1
```

```python
print(f"Decision Tree Classifier tahmini  = {grid_search.predict(np.array([0.51,0.65,6,270,5,0,1,6,2]).reshape(1, 9))}")
```

```python
#KMeans Clustering metodu
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
import numpy as np
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.show()
```
